### Столбчатая диаграмма для категориальной переменной Class:

Столбчатые диаграммы для категориальных переменных

```
# Выберем самые показательные категориальные переменные
key_categorical = ['Gender', 'BMI Category', 'Sleep Disorder', 'Occupation']

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.ravel()

for i, col in enumerate(key_categorical):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Распределение {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Количество')
    # Поворачиваем подписи для Occupation для лучшей читаемости
    if col == 'Occupation':
        axes[i].tick_params(axis='x', rotation=45)

    # Добавляем аннотации с количеством
    for p in axes[i].patches:
        axes[i].annotate(f'{p.get_height()}',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### Боксплоты для числовых переменных

```
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.ravel()

numeric_for_boxplot = ['Sleep Duration', 'Age', 'Heart Rate', 'Daily Steps']

for i, col in enumerate(numeric_for_boxplot):
    sns.boxplot(data=df, y=col, ax=axes[i])
    axes[i].set_title(f'Боксплот {col}')
    axes[i].set_ylabel(col)

    # Добавляем аннотации с основными статистиками
    stats = df[col].describe()
    axes[i].text(0.7, 0.95, f'Медиана: {stats["50%"]:.1f}\nIQR: {stats["75%"]-stats["25%"]:.1f}',
                transform=axes[i].transAxes, bbox=dict(boxstyle="round", facecolor='wheat'))

plt.tight_layout()
plt.show()
```

### Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:

```
# Сначала убедимся, что работаем только с числовыми переменными
print("Типы данных в датасете:")
print(df.dtypes)

# Выделим числовые колонки
numeric_columns = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level',
                   'Stress Level', 'Heart Rate', 'Daily Steps']

# Создадим корреляционную матрицу
correlation_matrix = df[numeric_columns].corr()

# Визуализация тепловой карты
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.2f', linewidths=0.5)
plt.title('Корреляционная матрица числовых переменных', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```
